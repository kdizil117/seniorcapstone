```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

#Wage data
wage_df = hourly_earnings_df.copy()
wage_df['Label'] = pd.to_datetime(wage_df['Label'], errors='coerce')
wage_df.dropna(subset=['Label'], inplace=True)
wage_df['year'] = wage_df['Label'].dt.year

# Convert hourly to annual wage
wage_df['annual_income'] = wage_df['Value'] * 2080

# Average per year
annual_wage = wage_df.groupby('year')['annual_income'].mean().reset_index()

# Housing data 
housing_df = Housing_prices_df.copy()
housing_df['year'] = housing_df['date_column'].dt.year
annual_housing = housing_df.groupby('year')['MSPUS'].mean().reset_index()

# Merge both on year
merged_df = pd.merge(annual_housing, annual_wage, on='year', how='inner')

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Plot raw home price vs income 
fig, ax = plt.subplots(figsize=(16, 8))

# Plot lines with style
ax.plot(merged_df['year'], merged_df['MSPUS'], label='Average Home Price', linewidth=2.5, color='#1f77b4')
ax.plot(merged_df['year'], merged_df['annual_income'], label='Average Annual Income (40hrs/week)', linewidth=2.5, color='#ff7f0e')

# Titles and labels
ax.set_title("U.S. Housing Prices vs Income Over Time", fontsize=20, weight='bold')
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("USD (Dollars)", fontsize=14)

# Format Y axis with commas
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f"${x:,.0f}"))

# Format ticks
ax.tick_params(axis='both', which='major', labelsize=12)

# Grid and legend
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend(loc='upper left', fontsize=13, frameon=False)

# Optional: highlight 2025
ax.axvline(x=2025, color='gray', linestyle=':', linewidth=1)
ax.text(2025 + 0.5, ax.get_ylim()[1] * 0.95, '2025', color='gray', fontsize=11)

plt.tight_layout()
plt.show()


```

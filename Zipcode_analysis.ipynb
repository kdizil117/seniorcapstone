```python
import os
import pyodbc

folder_path = '/home/kalle/Documents/seniorcapstone/seniorcapstone/State-zip-code-GeoJSON-master'

conn_str = (
    'DRIVER={ODBC Driver 18 for SQL Server};'
    'SERVER=localhost,1433;'
    'DATABASE=location_US_Json;'
    'UID=sa;'
    'PWD=SuperK_0606$;'
    'Encrypt=yes;TrustServerCertificate=yes;'
)

try:
    conn = pyodbc.connect(conn_str)
    cursor = conn.cursor()
    print("Connected to DB:", conn.getinfo(pyodbc.SQL_DATABASE_NAME))

    for filename in os.listdir(folder_path):
        if filename.lower().endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as f:
                json_text = f.read()

            try:
                cursor.execute(
                    "INSERT INTO US_Zip_Json (JsonContent) VALUES (?)",
                    json_text
                )
                print(f'Inserted: {filename}')
                conn.commit()
            except Exception as e:
                print(f'Error inserting {filename}: {e}')
    
    cursor.close()
    conn.close()

except Exception as e:
    print("Database connection failed:", e)


```
